```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
plt.show()
```

```python
z = np.linspace(0,1,100)
x = z * np.sin(20*z)
y = z * np.cos(20*z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot(x,y,z,color='red')

ax.set(title='Contoh 3D line plot',xlabel = 'sumbu X',ylabel = 'sumbu Y',zlabel = 'sumbu Z')
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
n = np.linspace(-3.0,3.0,100)

x,y = np.meshgrid(n,n)
z = np.sqrt(x**2+y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

```python
plt.contour(x,y,z)
plt.colorbar()

plt.title('Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.contourf(x,y,z)
plt.colorbar()

plt.title('Filled Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
n = np.linspace(-3.0,3.0,100)

x,y = np.meshgrid(n,n)
z = np.sqrt(x**2+y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_wireframe(x,y,z,color='orange')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Wireframe')

plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x,y,z,cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Surface Plot')

plt.show()
```
