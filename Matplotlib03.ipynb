```python
#mengimport module yang digunakan 
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
#mencetak versi matplotlib dan numpy yang sedang digunakan
print(matplotlib.__version__)
print(np.__version__)
```

```python
#membuat array dengan start point 0 hingga end point 2.0 dengan interval 0.01
#starting point bersifat inclusive sedangkan ending point bersifat exclusive
x = np.arange(0.0,2.0,0.01)
x
```

```python
# dalam trigonometri nillai 2 * pi * x adalah radian
# membentuk array yang berisi nilai sinus dari data point sebelumnya
s = np.sin(2*np.pi*x)
s
```

```python
# melakukan plotting dengan OO Style 
# variabel pertama akan menampung figure dan variabel kedua akan menampung axes
fig,ax = plt.subplots()
#plotting akan dilakukan dengan 'x' sebagai sumbu x dan 's' sebagai sumbu y
ax.plot(x,s)
ax.set(xlabel ='nilai x',
      ylabel  ='nilai sin(x)',
      title   ='Visualisasi data sin')
# ax.grid() merupakan fungsi untuk menampilkan kotak-kotak samar pada area plotting
ax.grid()
plt.show()
```

```python
#plotting dengan cara pyplot style
plt.plot(x,s)
plt.xlabel('nilai x')
plt.ylabel('nilai sin(x)')
plt.title('Visualisasi nilai sin')
plt.grid()
```

```python
#linspace berbeda dengan arange dimana jika pada linspace 2 parameter pertama bersifat inklusif 
#pada arange parameter pertama bersifat inklusiv sedangkan parameter kedua bersifat eksklusif
#parameter ketiga pada linspace mengindikasikan jumlah datapoint yang dibentuk
x1 = np.linspace(0.0,5.0,100)
x2 = np.linspace(0.0,2.0,200)

#memanggil fungsi cos untuk setiap data point 'x1' dan 'x2'
y1 = np.cos(2*np.pi*x1)
y2 = np.cos(2*np.pi*x2)
```

```python
x1
```

```python
# parameter 2,1 mengindikasikan kita ingin membuat plotting dengan bentuk 2 baris dan 2 kolom
# fig digunakan untuk menampung figure sedangkan ax1 dan ax2 digunakan untuk menampung dua buah axes
fig,(ax1,ax2) = plt.subplots(2,1)
#subtitle() digunakan untuk mengatur judul plotting
fig.suptitle('Multiple Subplots')

#parameter ketiga mengindikasikan hasil plotting akan berwarna merah (red) dan akan berbentuk bulatan (-o)
ax1.plot(x1,y1,'ro-')
ax1.set_ylabel('Nilai cos$(x_1)$')

#parameter ketiga mengindikasikan hasil plotting akan berwarna hijau (green) dan akan berbentuk bulatan (-)
ax2.plot(x2,y2,'g.-')
ax2.set_ylabel('Nilai cos$(x_2)$')

ax2.set_xlabel('Nilai x')
plt.show()
```

```python
#parameter subplots yang pertama dan kedua mengindikasikan jumlah baris dan kolom tempat kita melakukan plotting
#sedangkan parameter ketiga yang bernilai 1 mengindikasikan bahwa kita ingin mengendalikan 1 akses dari 2 akses yang kita miliki
plt.subplot(2,1,1)
plt.plot(x1,y1,'ro-')
plt.title('Multiple Subplots')
plt.ylabel('Nilai cos$(x_1)$')

#melakukan plotting pada axes kedua
plt.subplot(2,1,2)
plt.plot(x2,y2,'g.-')
plt.ylabel('Nilai cos$(x_2)$')
plt.xlabel('Nilai $x$')

plt.show()
```

```python
#mengimport module yang digunakan 
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
#mencetak versi matplotlib dan numpy yang sedang digunakan
print(matplotlib.__version__)
print(np.__version__)
```

```python
mu,sigma = 100,15 #nilai mean dan nilai standard deviation

#pembentukan 10000 angka yang membentuk distribusi normal
x = mu + sigma * np.random.randn(10000) #normal distribution
x
```

```python
#melihat jumlah data yang tersimpan dalam variabel x
x.shape
```

```python
#fungsi .hist() memerlukan setidaknya dua parameter, parameter pertama mengindikasikan dataset sedangkan parameter kedua 
#merupakan bins yang mengindikasikan jumlah pengelompokan data 
#parameter alpha digunakan untuk mengatur transparansi
plt.hist(x,
        bins=50,
        facecolor='g',
        alpha=0.75)

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogram')

plt.text(45,500,r'$\mu=100,\ \sigma=15$')
plt.grid()
plt.show()
```

```python
fig,ax = plt.subplots()
ax.hist(x,
       bins=50,
       facecolor='r',
       alpha = 0.75)
ax.set_xlabel = ('Sumbu X')
ax.set_ylabel = ('Sumbu Y')
ax.set_title ('Contoh Histogram')

ax.text(45,500,r'$\mu=100,\ \sigma=15$')
ax.grid()
plt.show()
```
