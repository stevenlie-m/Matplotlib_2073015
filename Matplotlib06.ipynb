```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
np.random.seed(2)
data = np.random.normal(loc=100,scale=10,size=200)
data
```

```python
plt.boxplot(data,labels=['Data'])

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()
plt.show()
```

```python
plt.boxplot(data,labels=['Data'],notch = True)

plt.title('Notched Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.boxplot(data,labels=['Data'],showfliers=True,flierprops={'markerfacecolor':'r','marker':'d'})

plt.title('Pengaturan Marker')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.boxplot(data,labels=['Data'],vert=False)
plt.title('Horisontal Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
np.random.seed(2)
data = np.random.normal(loc=100,scale=10,size=200)
data
```

```python
plt.violinplot(data)
plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data,
              showextrema=False,
              showmeans = False,
              showmedians = False,
              quantiles = None)
plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data , vert = False)
plt.title('Horisontal Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
x = np.linspace(1,10,25)
x
```

```python
fig,ax1 = plt.subplots()
ax1.plot(x,np.exp(x),'bs-',label='exp')
ax1.setx_label = ('Sumbu X')
ax1.sety_label = ('esp')

ax2 = ax1.twinx()
ax2.plot(x,np.log(x),'ro-',label='log')
ax1.sety_label = ('log')

fig.suptitle('Contoh Twin Axes')
fig.legend(loc = 'upper left')

plt.show()
```

```python
x = np.random.randint(16,40,size=30)
x
```

```python
def C2F(celcius=0):
    return (celcius*1.8)+32

def konversi_sumbu(ax1):
    y1,y2=ax1.get_ylim()
    ax2.set_ylim(C2F(y1),C2F(y2))
    ax2.figure.canvas.draw()
    
fig,ax1= plt.subplots()
ax2 = ax1.twinx()

ax1.callbacks.connect('y_lim changed',konversi_sumbu)
ax1.plot(x)

ax1.set_xlabel('Hari')
ax1.set_ylabel('Celcius')
ax2.set_ylabel('Fahrenheit')

fig.suptitle('Temperature Udara')
plt.show()
```
